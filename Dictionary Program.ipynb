## 2. Access a value from a dictionary.


```python
# Question: Given the dictionary {"name": "Bob", "age": 30}, print the age.
```

```python
person={"name":"Bob","age":30}
print(person["age"])
```

## 3. Add a new key-value pair to a dictionary.

```python
# Question: Add a key "city" with value "New York" to the dictionary person.
```

```python
person={"name":"Bob","age":30}
person["city"]="New york"
print(person)
```

```python
# Question: Change the age of "Alice" to 28 in the given dictionary.
```

```python
person={"name":"salman khan","age":18}
person["age"]=19
print(person)
```

## 5. Delete a key from a dictionary.
